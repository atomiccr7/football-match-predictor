```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("tips.csv")
```

```python
dataset.sample(3)
```

```python
dataset.info()
```

```python
data = dataset.select_dtypes(["float64","int64"]).corr()
```

```python
data_c = dataset.select_dtypes(["float64","int64"]).cov()
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(data,annot = True)
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(data_c,annot=True)
plt.show()
```
