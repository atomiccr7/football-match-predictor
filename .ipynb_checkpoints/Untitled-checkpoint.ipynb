```python
import scipy.stats as st
import numpy as np
```

```python
s_x = 90
p_u = 82
p_std = 20
n = 81
ap = 0.05
```

```python
z_cal = (s_x-p_u)/(p_std/np.sqrt(n))
z_cal
```

```python
z_table = st.norm.ppf(1-ap)
z_table
```

```python
if z_table<z_cal:
    print("ha is right")
else:
    print("h0 is right")
```
